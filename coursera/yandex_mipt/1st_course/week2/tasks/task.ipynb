```python
file = open('cats.txt')
```

```python
file = file.readlines()
```

```python
file
```

```python
for i in range(len(file)):
    file[i] = file[i].lower().strip()
```

```python
file
```

```python
print(file[0])
```

```python
import re

for i in range(len(file)):
    file[i] = re.split('[^a-z]', file[i])
```

```python
file
```

```python
for i in range(len(file)):
    for word in file[i]:
        if word == '':
            file[i].remove(word)
```

```python
file[0]
```

```python
file
```

```python
words = []

for line in file:
    for word in line:
        words.append(word)
```

```python
words
```

```python
unique_words = set(words)
```

```python
len(words)
```

```python
len(unique_words)
```

```python
words_dict = {}

i = 0
for unique_word in unique_words:
    words_dict[unique_word] = i
    i += 1
```
